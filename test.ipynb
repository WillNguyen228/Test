```python
# Colab Notebook: Simple Sine Plot

# Install dependencies (only needed if not already installed)
%pip install -q numpy matplotlib

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 500)  # 0 to 2Ï€
y = np.sin(x)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Simple Sine Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```
